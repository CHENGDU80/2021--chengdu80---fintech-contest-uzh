```python
import numpy as np
import pandas as pd

company_ar_alterstockinfo = pd.read_csv('company_ar_alterstockinfo.csv', dtype = object)

factor1 = []
df_factor1 = pd.DataFrame()

for name in company_ar_alterstockinfo['entid'].unique():
    subdata = company_ar_alterstockinfo[company_ar_alterstockinfo['entid'] == name]
    transsum = 0
    for i in range(len(subdata)):
        trans = float(subdata.iloc[i,4])-float(subdata.iloc[i,3])
        if trans < 0:
            trans = trans * (-2) # double weight for negatives
        transsum += trans
    factor1.append(transsum)
    
df_factor1['entid'] = pd.Series(company_ar_alterstockinfo['entid'].unique())
df_factor1['weighted_sum_trans_equity'] = pd.Series(factor1)
print(df_factor1)
```

```python
company_ar = pd.read_csv('company_ar.csv', dtype = object)
factor2 = []
df_factor2 = pd.DataFrame()
for name in company_ar['entid'].unique():
    subdata = company_ar[company_ar['entid'] == name]
    femaleratio = []
    for i in range(len(subdata)):
        femaleratio.append(np.divide(float(subdata.iloc[i,4]),float(subdata.iloc[i,3])))
    a = np.array(femaleratio)
    femaleratioavg = np.mean(a[a<10])
    factor2.append(femaleratioavg)
df_factor2['entid'] = pd.Series(company_ar['entid'].unique())
df_factor2['female_male_ratio'] = pd.Series(factor2)
print(df_factor2)
```

```python
company_ar_assetsinfo = pd.read_csv('company_ar_assetsinfo.csv', dtype = object)
company_ar_assetsinfo
df_factor3 = pd.DataFrame()
factor31 = []
factor32 = []
factor33 = []
factor34 = []
factor35 = []
for name in company_ar_assetsinfo['entid'].unique():
    subdata = company_ar_assetsinfo[company_ar_assetsinfo['entid'] == name]
    gross_assets = []
    liability_assets_ratio = []
    oper_inc_net_inc_ratio = []
    main_busi_net_inc_ratio = []
    profit_net_inc_ratio = []
    for i in range(len(subdata)):
        gross_assets.append(float(subdata.iloc[i,3]))
        liability_assets_ratio.append(np.divide(float(subdata.iloc[i,4]),float(subdata.iloc[i,3])))
        oper_inc_net_inc_ratio.append(np.divide(float(subdata.iloc[i,5]),float(subdata.iloc[i,8])))
        main_busi_net_inc_ratio.append(np.divide(float(subdata.iloc[i,6]),float(subdata.iloc[i,8])))
        profit_net_inc_ratio.append(np.divide(float(subdata.iloc[i,7]),float(subdata.iloc[i,8])))
    factor31.append(pd.Series(gross_assets).replace([np.inf,-np.inf],np.nan).mean())
    factor32.append(pd.Series(liability_assets_ratio).replace([np.inf,-np.inf],np.nan).mean())
    factor33.append(pd.Series(oper_inc_net_inc_ratio).replace([np.inf,-np.inf],np.nan).mean())
    factor34.append(pd.Series(main_busi_net_inc_ratio).replace([np.inf,-np.inf],np.nan).mean())
    factor35.append(pd.Series(profit_net_inc_ratio).replace([np.inf,-np.inf],np.nan).mean())
df_factor3['entid'] = pd.Series(company_ar_assetsinfo['entid'].unique())
df_factor3['gross_assets'] = pd.Series(factor31)
df_factor3['liability_assets_ratio'] = pd.Series(factor32)
df_factor3['oper_inc_net_inc_ratio'] = pd.Series(factor33)
df_factor3['main_busi_net_inc_ratio'] = pd.Series(factor34)
df_factor3['profit_net_inc_ratio'] = pd.Series(factor35)
print(df_factor3)
```

```python
company_ar_nz = pd.read_csv('company_ar_nz.csv', dtype = object)
df_factor4 = pd.DataFrame()
factor41 = []
factor42 = []
for name in company_ar_nz['entid'].unique():
    subdata = company_ar_nz[company_ar_nz['entid'] == name]
    farmer_ratio = []
    new_farmer_ratio = []
    for i in range(len(subdata)):
        farmer_ratio.append(np.divide(float(subdata.iloc[i,2]),float(subdata.iloc[i,1])))
        new_farmer_ratio.append(np.divide(float(subdata.iloc[i,3]),float(subdata.iloc[i,4])))
    factor41.append(pd.Series(farmer_ratio).replace([np.inf,-np.inf],np.nan).mean())
    factor42.append(pd.Series(new_farmer_ratio).replace([np.inf,-np.inf],np.nan).mean())
df_factor4['entid'] = pd.Series(company_ar_nz['entid'].unique())
df_factor4['farmer_ratio'] = pd.Series(factor41)
df_factor4['new_farmer_ratio'] = pd.Series(factor42)
print(df_factor4)
```

```python
from scipy import stats
df_factor8 = pd.DataFrame()
tax_company = pd.read_csv('tax_company.csv', dtype = object)
tax_company = tax_company[['entid','registertype','economic_type']]
register = []
economic = []
for name in tax_company['entid'].unique():
    subdata = tax_company[tax_company['entid'] == name]
    register_type_value = float(stats.mode(np.array(subdata['registertype']))[0])
    economic_type_value = float(stats.mode(np.array(subdata['economic_type']))[0])
    register.append(register_type_value)
    economic.append(economic_type_value)
df_factor8['entid'] = tax_company['entid'].unique()
df_factor8['register_type'] = pd.Series(register)
df_factor8['economic_type'] = pd.Series(economic)
df_factor8
```

```python
tax_year = pd.read_csv('tax_year.csv', dtype = object)
tax_year = tax_year[['entid','profit_loss','fair_value_change','operation_profit','total_profit','assets_re']]
df_factor9 = pd.DataFrame()
df_factor9['entid'] = tax_year['entid'].unique()
factor91 = []
factor92 = []
factor93 = []
factor94 = []
factor95 = []
for name in tax_year['entid'].unique():
    pnl_dummy = []
    fair_value_change = []
    oper_total_profit_ratio = []
    total_profit = []
    asset_re = []
    subdata = tax_year[tax_year['entid'] == name]
    for i in range(len(subdata)):
        pnl_dummy.append(np.sign(float(subdata.iloc[i,1])))
        fair_value_change.append(float(subdata.iloc[i,2]))
        oper_total_profit_ratio.append(np.divide(float(subdata.iloc[i,3]),float(subdata.iloc[i,4])))
        total_profit.append(float(subdata.iloc[i,4]))
        asset_re.append(float(subdata.iloc[i,5]))
    factor91.append(pd.Series(pnl_dummy).replace([np.inf,-np.inf],np.nan).mean())
    factor92.append(pd.Series(fair_value_change).replace([np.inf,-np.inf],np.nan).mean())
    factor93.append(pd.Series(oper_total_profit_ratio).replace([np.inf,-np.inf],np.nan).mean())
    factor94.append(pd.Series(total_profit).replace([np.inf,-np.inf],np.nan).mean())
    factor95.append(pd.Series(asset_re).replace([np.inf,-np.inf],np.nan).mean())
df_factor9['pnl_dummy'] = pd.Series(factor91)
df_factor9['fair_value_change'] = pd.Series(factor92)
df_factor9['oper_total_profit_ratio'] = pd.Series(factor93)
df_factor9['total_profit'] = pd.Series(factor94)
df_factor9['asset_re'] = pd.Series(factor95)
df_factor9
```

```python
from scipy.stats import norm
def MertonModel(debt, asset, mu = 0.1, T = 1, sigma):
    a = (np.log(debt)-np.log(asset)-mu*T-0.5*(sigma**2)*T)/(sigma*np.sqrt(T))
    return norm.cdf(a)
```
