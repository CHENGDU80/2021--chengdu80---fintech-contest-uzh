```python
import pandas as pd
from sklearn.model_selection import train_test_split
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score
from sklearn.model_selection import cross_val_score


train_data = pd.read_csv( 'data/train_dataset_final.csv') #读取训练数据
test_data = pd.read_csv( 'data/test_dataset25-1.csv')  #读取测试数据

train_y = train_data.CaseType #获取trian_y
train_X = train_data.drop(['CaseType'], axis=1).select_dtypes(exclude=['object'])#只使用数字预测器

test_y = test_data.CaseType #获取trian_y
test_X = test_data.drop(['CaseType'], axis=1).select_dtypes(exclude=['object'])#只使用数字预测器

# 4.空值处理，默认方法：使用特征列的平均值进行填充
my_imputer = SimpleImputer()
train_X = my_imputer.fit_transform(train_X)
test_X = my_imputer.transform(test_X)
```

```python

print("LGB test")
clf = lgb.LGBMClassifier(
        boosting_type='gbdt', num_leaves=55, reg_alpha=0.0, reg_lambda=1,
        max_depth=15, n_estimators=6000, objective='multiclass',
        subsample=0.8, colsample_bytree=0.8, subsample_freq=1,
        learning_rate=0.06, min_child_weight=1, random_state=20, n_jobs=4, 
#         eval_metric  = 'logloss',
#         early_stopping_rounds=150,
    )
clf.fit(train_X, train_y,eval_set = (test_X, test_y))
pre=clf.predict(test_X)
```

```python
test_accuracy = accuracy_score(test_y, pre)

print("Test Accuracy: %.2f%%" % (test_accuracy * 100.0))

print("recall_score=",recall_score(test_y, pre, average='micro'))
print("precision_score=",precision_score(test_y, pre, average='micro'))
```

```python

print(cross_val_score(clf,train_X,train_y,cv=5))
```
